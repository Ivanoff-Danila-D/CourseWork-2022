```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def image_triangle(side_1, side_2, side_3, ax):
    ax.plot(side_1.x, side_1.y, color='black')
    ax.plot(side_2.x, side_2.y, color='black')
    ax.plot(side_3.x, side_3.y, color='black')
    
    ax.scatter(side_1.x[0], side_1.y[0], color='black')
    ax.scatter(side_2.x[0], side_2.y[0], color='black')
    ax.scatter(side_3.x[0], side_3.y[0], color='black')
```

```python
def image_tails(tail_1, tail_2, tail_3, ax):
    ax.plot(tail_1.x, tail_1.y, color='red')
    ax.plot(tail_2.x, tail_2.y, color='red')
    ax.plot(tail_3.x, tail_3.y, color='red')
```

```python
def image_final_triangle(side_1_new, side_2_new, side_3_new, ax):
    ax.plot(side_1_new.x, side_1_new.y, color='blue')
    ax.plot(side_2_new.x, side_2_new.y, color='blue')
    ax.plot(side_3_new.x, side_3_new.y, color='blue')
```

```python
side_1 = pd.read_csv('side_1.txt', names=['x', 'y'])
side_2 = pd.read_csv('side_2.txt', names=['x', 'y'])
side_3 = pd.read_csv('side_3.txt', names=['x', 'y'])

tail_1 = pd.read_csv('tail_1.txt', names=['x', 'y'])
tail_2 = pd.read_csv('tail_2.txt', names=['x', 'y'])
tail_3 = pd.read_csv('tail_3.txt', names=['x', 'y'])

side_1_new = pd.read_csv('side_1_new.txt', names=['x', 'y'])
side_2_new = pd.read_csv('side_2_new.txt', names=['x', 'y'])
side_3_new = pd.read_csv('side_3_new.txt', names=['x', 'y'])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
image_triangle(side_1, side_2, side_3, ax)
image_tails(tail_1, tail_2, tail_3, ax)
image_final_triangle(side_1_new, side_2_new, side_3_new, ax)
plt.show()
```
